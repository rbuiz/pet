# Парсинг страниц сайта с пагинацией.

## Часть 2. Загрузка HTML-кода из файлов и работа с ним. 

### 1. Подключение библиотек.

```python
import bs4
import pandas as pd
```

### 2. Цикл для загрузки и обработки HTML-кода 

при подготовки к парсингу в исходном коде одной из спарсимуемых страниц подбираем теги с интересующей информацией, записываем их классы.

**в цикле выполняются операции:**
- загрузка кода из полученных в 1-ой части файлов;
- получение BeautifulSoup для парсинга;
- получение текстового содержимого из интересующих тегов.
- обработка полученных текстовых значений - в данном случае из значения цены убираем знак доллара и приводим к типу чисел с плавающей точкой. Из названий товара убираем пробелы перед строкой и после неё;
- добавление полученных значений в список.

```python
l = []

for i in range(1, 8):
    txt = open('pag/soup_'+str(i)+'.txt','r').read()
    b = bs4.BeautifulSoup(txt, 'html.parser')
    dv = b.select('div.col-md-6.mb-4')
    for d in range(len(dv)):
        items = dv[d].select('h4 a')
        prises = dv[d].select('h5')
        ld = [items[0].text.strip(), float(prises[0].text.strip().replace('$', ''))]
        l.append(ld)
```

### 3. Конвертация полученного списка в DataFrame.

```python
df = pd.DataFrame(data = l, columns=[['item', 'price']])
display(df)
df.info()
```

### 4. Запись полученной таблицы в exel и csv файлы на локальном диске.

```python
df.to_csv('prices.csv', encoding='utf-8', index=False)

writer = pd.ExcelWriter('prices.xlsx')
df.to_excel(writer)
writer.save()
```
